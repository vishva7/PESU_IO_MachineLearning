### Importing Libraries and Modules

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_absolute_error
```

### Importing Dataset

```python
dataset = pd.read_csv('IMDb_All_Genres_etf_clean1.csv')
dataset
```

## Data Preprocessing

### Dropping movies with unknown grosses

```python
dataset = dataset.replace('Gross Unkown', np.nan)
dataset = dataset.dropna(subset=['Total_Gross'])
```

```python
dataset['Total_Gross'] = dataset['Total_Gross'].str.replace('M', '')
dataset['Total_Gross'] = dataset['Total_Gross'].str.replace('$', '')
dataset['Total_Gross'] = pd.to_numeric(dataset['Total_Gross'], errors='coerce').fillna(0).astype(float)
```

### Splitting the actor column and binary encoding it

```python
dataset[['Actor 1', 'Actor 2','Actor 3','Actor 4']] = dataset['Actors'].str.split(',',expand=True)
dataset = dataset.drop('Actors', axis=1)
dataset
```

```python
dataset = dataset.loc[:,['Movie_Title','Director','Year','Rating','Runtime(Mins)','Actor 1','Actor 2','Actor 3','Actor 4','main_genre','side_genre','Total_Gross']]
dataset
```

```python
from category_encoders.binary import BinaryEncoder
columns = ["Actor 1", "Actor 2", "Actor 3", "Actor 4"]
encoder = BinaryEncoder(cols=columns, drop_invariant=False)
dataset=encoder.fit_transform(dataset)
dataset
```

### Binary Encoding the Main Genre

```python
from category_encoders.binary import BinaryEncoder
encoder = BinaryEncoder(cols=['main_genre'], drop_invariant=False)
dataset=encoder.fit_transform(dataset)
dataset
```

### Splitting the Side Genres and binary encoding them

```python
from category_encoders.binary import BinaryEncoder
dataset[['Side_Genre1','Side_Genre2']] = dataset['side_genre'].str.split(',',expand=True)
dataset = dataset.drop('side_genre', axis=1)
dataset
```

```python
from category_encoders.binary import BinaryEncoder
columns = ["Side_Genre1", "Side_Genre2"]
encoder = BinaryEncoder(cols=columns, drop_invariant=False)
dataset=encoder.fit_transform(dataset)
dataset
```

```python
gross = dataset.pop('Total_Gross')
dataset.insert(66, 'Total_Gross', gross)
dataset
```

### Taking independent and dependent variables

```python
X = dataset.iloc[:, 2:66].values
Y = dataset.iloc[:, 66].values
X
```

```python
Y
```

### Splitting into test and training models

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Regression

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
print("Linear Regression score:", lin_reg.score(X_test,y_test))
```

```python
lin_predictions = lin_reg.predict(X_test)
MAE = mean_absolute_error(y_test, lin_predictions)
print('Linear Regression validation MAE =', MAE)
```

### Polynomial Regression

```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

max = 0
max_terms = [0,0]

for i in range(1,11):
    pca = PCA(n_components = i)
    X_poly_train = pca.fit_transform(X_train)
    X_poly_test = pca.transform(X_test)

    for deg in range(2,6):
        poly_reg = PolynomialFeatures(degree = deg)
        X_poly = poly_reg.fit_transform(X_poly_train)
        lin_reg_2 = LinearRegression()
        lin_reg_2.fit(X_poly, y_train)
        score = lin_reg_2.score(PolynomialFeatures(degree = deg).fit_transform(X_poly_test), y_test)
        if score > max:
            max = score
            max_terms[0] = i
            max_terms[1] = deg
        # print(score)

print("Polynomial Regression", max, max_terms)
```

```python
pca = PCA(n_components = max_terms[0])
X_poly_train = pca.fit_transform(X_train)
X_poly_test = pca.transform(X_test)

poly_reg = PolynomialFeatures(degree = max_terms[1])
X_poly = poly_reg.fit_transform(X_poly_train)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y_train)

poly_predictions = lin_reg_2.predict(PolynomialFeatures(degree = max_terms[1]).fit_transform(X_poly_test))
MAE = mean_absolute_error(y_test, poly_predictions)
print('Polynomial Regression validation MAE =', MAE)
```

### Visualisation

```python
x = range(len(y_test))
plt.plot(x, y_test, linewidth=1, label="original")
plt.plot(x, poly_predictions, linewidth=1.1, label="predicted")
plt.title("Polynomial Regression")
plt.xlabel('y_test')
plt.ylabel('Box Office Predictions')
plt.legend(loc='best',fancybox=True, shadow=True)
plt.grid(True)
plt.show() 
```

### XGBoost

```python
import xgboost as xg

max = 0
n = 0

for i in range(5,20):
    xgb_r = xg.XGBRegressor(objective='reg:squarederror', booster='gbtree', n_estimators = i, seed = 42)
    xgb_r.fit(X_train, y_train)
    pred = xgb_r.predict(X_test)
    xgb_score = xgb_r.score(X_test, y_test)
    # print(xgb_score)
    if max < xgb_score:
        max = xgb_score
        n = i
print("Max XGBoost score:", max, "\nFor n_estimators =", n)
```

```python
MAE = mean_absolute_error(y_test, pred)
print('XGBoost MAE = ', MAE)
```

### Visualisation

```python
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, linewidth=1, label="original")
plt.plot(x_ax, pred, linewidth=1.1, label="predicted")
plt.title("y-test and y-predicted data")
plt.xlabel('X-axis')
plt.ylabel('Box Office')
plt.legend(loc='best',fancybox=True, shadow=True)
plt.grid(True)
plt.show()
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train,y_train)

score = model.score(X_test, y_test)
print("Random Forest Score:", score)
```

```python
predicted_prices = model.predict(X_test)
MAE = mean_absolute_error(y_test , predicted_prices)
print('Random forest validation MAE =', MAE)
```
